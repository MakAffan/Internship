# Seleniunm Exception Handling

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import re
import time
```

```python
url = "https://www.amazon.in/"
driver = webdriver.Chrome()
driver.get(url)
```

```python
val = input()
```

```python
search = driver.find_element(By.XPATH, '//div[@class="nav-search-field "]//input')
search.send_keys(val)
submit = driver.find_element(By.XPATH, '//div[@class="nav-search-submit nav-sprite"]//span')
submit.click()
time.sleep(3)
```

Question 2

```python
urls = []
start = 0
end = 2
for page in range(start,end):
    link = driver.find_elements(By.XPATH, '//h2[@class="a-size-mini a-spacing-none a-color-base s-line-clamp-4"]/a')
    for i in link:
        urls.append(i.get_attribute('href'))
    nextpage = driver.find_element(By.XPATH, '//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    nextpage.click()
    time.sleep(2)
```

```python
brand = []
price = []
delivery = []
avail = []
```

```python
for url in urls:
    driver.get(url)
    time.sleep(3)
    
    try:
        x = driver.find_element(By.XPATH, '//span[@class="a-size-large product-title-word-break"]')
        brand.append(x.text)
    
    except NoSuchElementException:
        brand.append('-')
    
    try:
        x = driver.find_element(By.XPATH, '//span[@class="a-price-whole"]')
        price.append(x.text)
    
    except NoSuchElementException:
        price.append('-')
        
    try:
        x = driver.find_element(By.XPATH, '//a[@class="a-link-normal"]')
        delivery.append(x.text)
    
    except NoSuchElementException:
        delivery.append('-')

    try:
        x = driver.find_element(By.XPATH, '//span[@class="a-size-medium a-color-success"]')
        avail.append(x.text)
    
    except NoSuchElementException:
        avail.append('-')
```

```python
df = pd.DataFrame({"Product Name":brand,"Price":price,"Delivery":delivery,"Availability":avail})
df
```

```python
url = "https://www.flipkart.com/"
driver = webdriver.Chrome()
driver.get(url)
```

```python
search = driver.find_element(By.XPATH, '//input[@class="Pke_EE"]')
search.send_keys('smartphone')
submit = driver.find_element(By.XPATH, '//button[@class="_2iLD__"]')
submit.click()
```

```python
urls = []
link = driver.find_elements(By.XPATH, '//a[@class="CGtC98"]')
for i in link:
    urls.append(i.get_attribute('href'))
```

```python
brand = []
colour = []
RAM = []
storage = []
pcamera = []
scamera = []
size = []
battery = []
price = []
```

```python
for url in urls:
    driver.get(url)
    time.sleep(2)
    
    try:
        x = driver.find_element(By.XPATH, '//span[@class="VU-ZEz"]')
        brand.append(x.text)
    
    except NoSuchElementException:
        brand.append('-')
        
    try:
        x = driver.find_element(By.XPATH, '//div[@class="WGBwfw"]//ul/li')
        colour.append(x.text)
    
    except NoSuchElementException:
        colour.append('-')

    try:
        x = driver.find_element(By.XPATH, '//li[@class="_7eSDEz"]')
        i = x.text.split('|')
        RAM.append(i[0])
        storage.append(i[1])
    
    except NoSuchElementException:
        RAM.append('-')
        storage.append('-')
    except IndexError:
        RAM.append('-')
        storage.append('-')
        
    try:
        x = driver.find_element(By.XPATH, '//div[@class="xFVion"]/ul/li[3]')
        c = x.text.split('|')
        pcamera.append(c[0])
        scamera.append(c[1])
    
    except NoSuchElementException:
        pcamera.append('-')
        scamera.append('-')
    except IndexError:
        pcamera.append('-')
        scamera.append('-')
        
    try:
        x = driver.find_element(By.XPATH, '//div[@class="xFVion"]/ul/li[2]')
        size.append(x.text)
    
    except NoSuchElementException:
        size.append('-')
        
    try:
        x = driver.find_element(By.XPATH, '//div[@class="xFVion"]/ul/li[4]')
        battery.append(x.text)
    
    except NoSuchElementException:
        battery.append('-')
        
    try:
        x = driver.find_element(By.XPATH, '//div[@class="Nx9bqj CxhGGd"]')
        price.append(x.text)
    
    except NoSuchElementException:
        price.append('-')
```

```python
df = pd.DataFrame({"Brand":brand,"Colour":colour,"RAM":RAM,"ROM":storage,"Primary Camera":pcamera,"Secondary Camera":scamera,"size":size,"Battery":battery,"Price":price})
df
```
