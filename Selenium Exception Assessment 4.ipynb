Question 1

```python
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver = webdriver.Chrome()
driver.get(url)
```

```python
rank = []
name = []
artist = []
upload = []
view = []
```

```python
r=1
names = driver.find_elements(By.XPATH, '//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[1]')
for i in names:
    name.append(i.text)
    rank.append(r)
    r=r+1
    
artists = driver.find_elements(By.XPATH, '//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[2]')
for i in artists:
    artist.append(i.text)
    
uploads = driver.find_elements(By.XPATH, '//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[3]')
for i in uploads:
    upload.append(i.text)
    
views = driver.find_elements(By.XPATH, '//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[4]')
for i in views:
    view.append(i.text)
```

```python
df = pd.DataFrame({"Rank":rank,"Name":name,"Uploader":artist,"Upload Date":upload,"View":view})
df
```

```python
url = "https://www.billboard.com/"
driver = webdriver.Chrome()
driver.get(url)
```

```python
charts = driver.find_element(By.XPATH, '/html/body/div[3]/header/div/div[1]/div/div/div[2]/div/nav/ul/li[1]/a')
driver.get(charts.get_attribute('href'))
t100 = driver.find_element(By.XPATH, '//span[@class="c-label  "]/a')
driver.get(t100.get_attribute('href'))
```

```python
song =[]
artist = []
last = []
peak = []
weeks = []
```

```python
songs = driver.find_elements(By.XPATH, '//li[@class="lrv-u-width-100p"]/ul/li[1]')
for i in songs:
    x = re.split("\n",i.text)
    song.append(x[0])
    artist.append(x[1])
    
last_week = driver.find_elements(By.XPATH, '//li[@class="lrv-u-width-100p"]/ul/li[4]')
for i in last_week:
    last.append(i.text)
    
peek_rank = driver.find_elements(By.XPATH, '//li[@class="lrv-u-width-100p"]/ul/li[5]')
for i in peek_rank:
    peak.append(i.text)
    
charts = driver.find_elements(By.XPATH, '//li[@class="lrv-u-width-100p"]/ul/li[6]')
for i in charts:
    weeks.append(i.text)
```

```python
df = pd.DataFrame({"Song":song,"Artist":artist,"Last Week":last,"Peak Pos.":peak,"Weeks on Chart":weeks})
df
```

```python
url = "https://archive.ics.uci.edu/"
driver = webdriver.Chrome()
driver.get(url)
```

```python
nav = driver.find_element(By.XPATH, '//a[@class="btn-primary btn"]')
driver.get(nav.get_attribute('href'))
```

```python
name = []
dtype = []
task = []
atype = []
inst = []
attr = []
year = []
```

```python
urls = []
link = driver.find_elements(By.XPATH, '//a[@class="link-hover link text-xl font-semibold"]')
for i in link:
    urls.append(i.get_attribute('href')) 
```

```python
for url in urls:
    driver.get(url)
    time.sleep(2)
    
    try:
        x = driver.find_element(By.XPATH, '//h1[@class="text-3xl font-semibold text-primary-content"]')
        name.append(x.text)
    except NoSuchElementException:
        name.append('-')
    
    try:
        x = driver.find_element(By.XPATH, '//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div/p')
        dtype.append(x.text)
    except NoSuchElementException:
        dtype.append('-')
        
    try:
        x = driver.find_element(By.XPATH, '//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[3]/p')
        task.append(x.text)
    except NoSuchElementException:
        task.append('-')
        
    try:
        x = driver.find_element(By.XPATH, '//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[4]/p')
        atype.append(x.text)
    except NoSuchElementException:
        atype.append('-')
        
    try:
        x = driver.find_element(By.XPATH, '//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[5]/p')
        inst.append(x.text)
    except NoSuchElementException:
        inst.append('-')
        
    try:
        x = driver.find_element(By.XPATH, '//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[6]/p')
        attr.append(x.text)
    except NoSuchElementException:
        attr.append('-')
        
    try:
        x = driver.find_element(By.XPATH, '//h2[@class="text-sm text-primary-content"]')
        year.append(x.text)
    except NoSuchElementException:
        year.append('-')
        
        
```

```python
df = pd.DataFrame({"Dataset Name":name,"Data Type":dtype,"Tasks":task,"Attribute Type":atype,"Instances":inst,"Features":attr,"Year":year})
df
```
